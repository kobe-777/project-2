```python
import pandas as pd
from pathlib import Path
```

```python
application_record_df = pd.read_csv(Path('Resources/application_record.csv'), index_col='ID')
application_record_df
```

```python
credit_record_df = pd.read_csv(Path('Resources/credit_record.csv'), index_col='ID')
credit_record_df
```

```python
df_application = application_record_df[application_record_df.index.isin(credit_record_df.index)]
df_credit = credit_record_df[credit_record_df.index.isin(application_record_df.index)]
```
